```python
#Task 5: Feature Engineering - Part 1
#5.1 Feature Normalization
# Normalize float type features using z-score normalization
float_columns = housing.select_dtypes(include=['float64']).columns
housing[float_columns] = (housing[float_columns] - housing[float_columns].mean()) / housing[float_columns].std()
```

```python
# Display the mean and standard deviation of the normalized features
print(housing[float_columns].mean())
print(housing[float_columns].std())
```

```python

# Normalize float type features using z-score normalization
float_columns = housing.select_dtypes(include=['float64']).columns
housing_zscore = housing.copy()
housing_zscore[float_columns] = (housing_zscore[float_columns] - housing_zscore[float_columns].mean()) / housing_zscore[float_columns].std()

# Display the mean and standard deviation of the normalized features
print("Z-score Normalization:")
print(housing_zscore[float_columns].mean())
print(housing_zscore[float_columns].std())
```

```python
from sklearn.preprocessing import MinMaxScaler

```

```python

# Normalize float type features using min-max scaling
scaler = MinMaxScaler()
housing_minmax = housing.copy()
housing_minmax[float_columns] = scaler.fit_transform(housing_minmax[float_columns])

# Display the min and max of the normalized features
print("\nMin-Max Scaling:")
print(housing_minmax[float_columns].min())
print(housing_minmax[float_columns].max())
```

```python
from sklearn.preprocessing import RobustScaler

```

```python
# Normalize float type features using robust scaling
robust_scaler = RobustScaler()
housing_robust = housing.copy()
housing_robust[float_columns] = robust_scaler.fit_transform(housing_robust[float_columns])

# Display the median and IQR of the normalized features
print("\nRobust Scaling:")
print(housing_robust[float_columns].median())
print(housing_robust[float_columns].quantile(0.75) - housing_robust[float_columns].quantile(0.25))
```

```python


# Display the mean and standard deviation of the normalized features
print(housing[float_columns].mean())
print(housing[float_columns].std())
```

```python
# Task 6: Feature Engineering - Part 2
# 6.1 Handling Categorical Variables

# Create dummy variables for the ocean_proximity column
housing = pd.get_dummies(housing, columns=['ocean_proximity'], drop_first=True)
```

```python

# Concatenate the normalized features, dummy variables, and income_cat column
final_features = housing_robust[float_columns.tolist() + housing_robust.filter(regex='^ocean_proximity_').columns.tolist() + ['income_cat']]

# Display the updated features DataFrame
print("\nUpdated Features DataFrame:")
print(final_features.head())
```

```python
# Task 7: Splitting the Data into Train and Test Set
# 7.1 Data Splitting
from sklearn.model_selection import train_test_split

# Split the features into training and test sets (80% train, 20% test)
train_set, test_set = train_test_split(housing, test_size=0.2, random_state=42)
```

```python
# Ensure the distribution of income_cat is similar in both sets
print(train_set['income_cat'].value_counts() / len(train_set))
print(test_set['income_cat'].value_counts() / len(test_set))
```

```python
# Drop the income_cat column from the training and test sets
train_set = train_set.drop('income_cat', axis=1)
test_set = test_set.drop('income_cat', axis=1)
```

```python

# Create labels for the training and test sets
train_labels = train_set['median_house_value']
train_set = train_set.drop('median_house_value', axis=1)
test_labels = test_set['median_house_value']
test_set = test_set.drop('median_house_value', axis=1)
```

```python
# Task 8: Training the ML Model (Random Forest Regressor)
# 8.1 Model Training
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
```

```python
# Train a RandomForestRegressor on the training data with specified hyperparameters
forest_reg = RandomForestRegressor(n_estimators=100, random_state=42)
forest_reg.fit(train_set, train_labels)
```

```python
# Calculate and display the R^2 score on the training set
train_r2 = forest_reg.score(train_set, train_labels)
print(f'Training R^2: {train_r2}')
```

```python
# Predict on the training set and calculate the RMSE
train_predictions = forest_reg.predict(train_set)
train_rmse = np.sqrt(mean_squared_error(train_labels, train_predictions))
print(f'Training RMSE: {train_rmse}')
```

```python
#Task 9: Evaluating the Model on the Test Set
#9.1 Model Evaluation

# Evaluate the model on the test set by calculating the R^2 score and RMSE
test_r2 = forest_reg.score(test_set, test_labels)
print(f'Test R^2: {test_r2}')
test_predictions = forest_reg.predict(test_set)
test_rmse = np.sqrt(mean_squared_error(test_labels, test_predictions))
print(f'Test RMSE: {test_rmse}')
```

```python
# Compare true values and predictions using a DataFrame
comparison = pd.DataFrame({'Actual': test_labels, 'Predicted': test_predictions})
print(comparison.head())

```

```python
# Calculate and display the mean absolute error (MAE)
test_mae = np.mean(np.abs(test_labels - test_predictions))
print(f'Test MAE: {test_mae}')
```

```python
#Task 10: Feature Importance
#10.1 Feature Importance Analysis python
# Extract and display the feature importances from the fitted model
feature_importances = forest_reg.feature_importances_
features = train_set.columns
importances_df = pd.DataFrame({'Feature': features, 'Importance': feature_importances})
importances_df = importances_df.sort_values(by='Importance', ascending=False)

```

```python

# Plot the feature importances as a horizontal bar plot
plt.figure(figsize=(10, 5))
sns.barplot(x='Importance', y='Feature', data=importances_df)
plt.title('Feature Importances')
plt.show()
```
